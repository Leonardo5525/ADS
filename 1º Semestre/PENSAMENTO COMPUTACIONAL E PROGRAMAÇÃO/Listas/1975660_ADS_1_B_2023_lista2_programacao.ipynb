```python
a = int(input('Escolha um número = '))
if a == 0:
    print('A equação NÃO é do 2º grau, ENCERRADO')
b = int(input('Escolha um número = '))
c = int(input('Escolha um número = '))

delta = b**2 - 4*a*c
if delta < 0:
    print(f' Delta igual a {delta}, a equação não possui raizes reais pois o DELTA é NEGATIVO, programa ENCERRADO')
elif delta == 0:    
    x = (-(b) + delta**(1/2))/(2*a)
    print(f'Delta é igual {delta}, raiz é igual a {x} ')
elif delta > 0:
    x1 = (-(b) + delta**(1/2))/(2*a)
    x2 = (-(b) - delta**(1/2))/(2*a)
    print(f'x1 é igual {delta}, raiz 1 é igual a {x1} ')
    print(f'x2 é igual {delta}, raiz 2 é igual a {x2} ')
```

```python
salario = float(input('Qual o valor do salário do funcionário? '))

por = 0.2

if salario <= 800:
    aumento = salario + (salario*por)
elif salario <= 1300:
    por = por - 0.5
    aumento = salario + (salario*por)
elif salario <= 2500:
    por = por - 0.1
    aumento = salario + (salario*por)
elif salario > 2500:
    por = por - 0.15
    aumento = salario + (salario*por)

print(f'Salário antes do aumento é igual a R${salario:.2f}')
print(f'O percentual de aumento foi de {por*100} %')
print(f'O valor do aumento foi de R$ {aumento - salario:.2f}')
print(f'O novo salario é de R$ {aumento:.2F}')
```
