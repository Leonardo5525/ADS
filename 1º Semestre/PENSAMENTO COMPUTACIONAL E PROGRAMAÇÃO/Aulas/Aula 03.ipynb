```python
print('Qual a área e perimetro do quadrado?')
base= int(input('lado = '))
altura= int(input('lado = '))


area = base * altura
perimetro = 2*(base + altura)

print(f'perimetro = {perimetro}')
print(f'area = {area}')
```

```python
print('Qual a área e perimetro do quadrado?')
a = int(input('lado = '))
b = int(input('lado = '))
c = int(input('lado = '))
d = int(input('lado = '))

area = a * b
area = a * d
area = b * a
area = b * c
perimetro = a + b + c + d

print(f'perimetro = {perimetro}')
print(f'area = {area}')
```
