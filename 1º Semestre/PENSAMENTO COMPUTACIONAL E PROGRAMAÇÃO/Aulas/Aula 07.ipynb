```python
h = int(input('Qual o horário que se encontra? '))

if h >= 18 and h <= 24:
    print ('Noturno')
elif h >= 12:
    print ('Vespertino')
elif h >= 6:
    print ('Matutino')
elif h >= 0 and h <= 5:
    print ('Noturno')
```

```python
h = int(input('Qual o horário que se encontra? '))

if h <= 24:
    print ('Noturno')
elif h <= 18:
    print ('Vespertino')
elif h < 12:
    print ('Matutino')
elif h < 6:
    print ('Noturno')
else:
    print('Esse horário não existe')
```
