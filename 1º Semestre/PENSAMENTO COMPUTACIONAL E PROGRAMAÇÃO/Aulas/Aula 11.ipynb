```python
meses = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']
temperaturas = []
i = 0
total = 0 

while i < 12:
    import random
    temp = random.uniform(10.0, 35.0)
    temperaturas.append(temp)
    total += temp
    
    i += 1

print('Meses e suas temperaturas')
i = 0
while i < 12:
    print(f' A temperatura no mês de {meses[i]}: {round(temperaturas[i],2)}')
    i += 1

print('-'*30)
media = total/12
print(f'Media da temperatura anual: {round(media,2)}')
print('-'*30)

print('Meses com a temperatura abaixo da média:')
i = 0
while i < len(temperaturas):
    if temperaturas[i] < media:
        menos = round(media - temperaturas[i],2)
        print(f'{i + 1} - {meses[i]} está {menos} abaixo da média anual')

    i += 1

print('-'*30)

print('Meses com a temperatura acima da média')
i = 0
while i < len(temperaturas):
    if temperaturas[i] > media:
        maior = round(temperaturas[i] - media,2)
        print(f'{i + 1} - {meses[i]} está {maior} acima da média anual')

    i += 1
```
